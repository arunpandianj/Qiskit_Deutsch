```python
# =========================================================
# 1️⃣ Custom Oracle Design - Deutsch Algorithm
# Function: f(x) = x XOR 1 (i.e., NOT(x))
# =========================================================

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Define oracle for f(x) = x XOR 1
def deutsch_oracle_xor1(qc):
    qc.x(0)     # Invert input qubit (x -> NOT x)
    qc.cx(0, 1) # Apply CNOT
    qc.x(0)     # Undo inversion

# Build circuit
qc1 = QuantumCircuit(2, 1)
qc1.x(1)
qc1.h([0, 1])
deutsch_oracle_xor1(qc1)
qc1.h(0)
qc1.measure(0, 0)

# Simulate
sim = AerSimulator()
compiled_qc = transpile(qc1, sim)
result = sim.run(compiled_qc).result()
counts = result.get_counts()

display(qc1.draw("mpl"))
plot_histogram(counts)
print("Measurement Result:", counts)
```

```python
# =========================================================
# 2️⃣ Bloch Sphere Visualization
# Visualize state after Hadamard and after Oracle
# =========================================================

from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# Step 1: After Hadamard
qc_h = QuantumCircuit(2)
qc_h.x(1)
qc_h.h([0, 1])
state_after_h = Statevector(qc_h)
plot_bloch_multivector(state_after_h)

# Step 2: After Oracle
qc_o = qc_h.copy()
deutsch_oracle_xor1(qc_o)
state_after_oracle = Statevector(qc_o)
plot_bloch_multivector(state_after_oracle)
```

```python
# =========================================================
# 3️⃣ Deutsch–Jozsa Extension (2 Input Qubits)
# =========================================================

from qiskit import QuantumCircuit

# Define oracle for balanced function: f(x1, x2) = x1 XOR x2
def oracle_2qubit_balanced(qc):
    qc.cx(0, 2)
    qc.cx(1, 2)

# Circuit for 2 input + 1 output qubit
qc3 = QuantumCircuit(3, 2)
qc3.x(2)
qc3.h([0, 1, 2])
oracle_2qubit_balanced(qc3)
qc3.h([0, 1])
qc3.measure([0, 1], [0, 1])

# Simulate
sim = AerSimulator()
compiled = transpile(qc3, sim)
result = sim.run(compiled).result()
counts = result.get_counts()

display(qc3.draw("mpl"))
plot_histogram(counts)
print("Measurement Result:", counts)
```

```python
# =========================================================
# 4️⃣ Noise Impact Simulation
# Compare ideal vs noisy simulation
# =========================================================

from qiskit_aer.noise import NoiseModel, depolarizing_error

# Create a simple noise model
noise_model = NoiseModel()
noise_model.add_all_qubit_quantum_error(depolarizing_error(0.05, 1), ['h', 'x'])
noise_model.add_all_qubit_quantum_error(depolarizing_error(0.1, 2), ['cx'])

# Run noisy simulation
noisy_sim = AerSimulator(noise_model=noise_model)
noisy_result = noisy_sim.run(transpile(qc1, noisy_sim)).result()
noisy_counts = noisy_result.get_counts()

# Compare results
plot_histogram([counts, noisy_counts], legend=["Ideal", "Noisy"])
```

```python
# =========================================================
# 5️⃣ Compare Classical vs Quantum Queries
# =========================================================

# Classical evaluation of f(x) = x XOR 1
def classical_f(x):
    return x ^ 1

# Classical method (needs 2 evaluations)
f0 = classical_f(0)
f1 = classical_f(1)
if f0 == f1:
    print("Classical result: CONSTANT (2 evaluations)")
else:
    print("Classical result: BALANCED (2 evaluations)")

# Quantum result (from Deutsch algorithm)
if list(counts.keys())[0] == '0':
    print("Quantum result: CONSTANT (1 query)")
else:
    print("Quantum result: BALANCED (1 query)")
```
