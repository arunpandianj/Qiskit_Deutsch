```python
# Imports and basic setup
import math, random, warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

# Qiskit imports
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from qiskit.quantum_info import Statevector
from qiskit_aer import Aer
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError

import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Import AerSimulator separately
from qiskit_aer import AerSimulator
```

```python
def dj_oracle_circuit(n, func_type="custom", custom_f=None):
    """
    Build Oracle for Deutsch-Jozsa with n input qubits and 1 output qubit.
    func_type: "constant_zero", "constant_one", "balanced_parity", "custom"
    For custom, provide custom_f(x_int)->0/1.
    """
    qc = QuantumCircuit(n+1, name="oracle")
    if func_type == "constant_zero":
        pass
    elif func_type == "constant_one":
        qc.x(n)
    elif func_type == "balanced_parity":
        for i in range(n):
            qc.cx(i, n)
    elif func_type == "custom":
        if custom_f is None:
            raise ValueError("custom_f must be provided for func_type='custom'")
        # For each input basis state x where f(x)==1, flip target controlled on equality to x.
        # This implementation uses mcx (may be transpiled if backend doesn't support directly).
        for x in range(2**n):
            if custom_f(x) == 1:
                # prepare controls: flip qubits where bit==0
                for i in range(n):
                    if ((x >> i) & 1) == 0:
                        qc.x(i)
                # multi-controlled X (target is n)
                qc.mcx(list(range(n)), n)
                # undo preparation
                for i in range(n):
                    if ((x >> i) & 1) == 0:
                        qc.x(i)
    else:
        raise ValueError("Unknown func_type")
    return qc
```

```python
# 1-qubit custom f(x) = x XOR 1
def f_single_xor1(x):
    return x ^ 1  # x in {0,1}

# 2-qubit custom f(x) = least-significant-bit XOR 1 (depends only on bit0)
def f_two_custom(x):
    bit0 = x & 1
    return bit0 ^ 1
```

```python
display(Markdown("## 1-qubit Deutsch-Jozsa with custom oracle f(x)=x XOR 1"))

# Explicitly import visualization functions
from qiskit.visualization import plot_bloch_multivector, plot_histogram

oracle1 = dj_oracle_circuit(1, func_type="custom", custom_f=f_single_xor1)
dj1 = deutsch_jozsa(1, oracle1)

display(Markdown("### Circuit (1-qubit DJ)"))
display(dj1.draw(output='text', fold=200)) # Changed output to 'text'

# Prepare statevector before oracle (input in H, output in |->)
prep = QuantumCircuit(2)
prep.x(1)
prep.h([0,1])
sv_prep = Statevector(prep)

display(Markdown("### Bloch BEFORE oracle (full state shown; input qubit is index 0)"))
plot_bloch_multivector(sv_prep.data)
plt.show()

# State after oracle (before final H on inputs)
full = QuantumCircuit(2)
full.x(1)
full.h([0,1])
full.append(oracle1.to_instruction(), [0,1])
sv_full = Statevector(full)

display(Markdown("### Bloch AFTER oracle (full state shown; input qubit is index 0)"))
plot_bloch_multivector(sv_full.data)
plt.show()

# Run statevector simulation for measurement counts
sv_backend = Aer.get_backend('aer_simulator_statevector')
t = transpile(dj1, sv_backend)
res = sv_backend.run(t).result()
counts = res.get_counts()
display(Markdown("### Measurement counts (1-qubit DJ)"))
print(counts)
display(Markdown("**Interpretation:** measure all zeros -> constant; otherwise balanced."))
is_constant = (list(counts.keys()) == ['0'] and counts.get('0',0) == sum(counts.values()))
display(Markdown(f"**Detected as constant?** {is_constant}"))
```

```python
# 2-qubit Deutsch-Jozsa circuits
oracle2_const = dj_oracle_circuit(2, func_type="constant_zero")
dj2_const = deutsch_jozsa(2, oracle2_const)

oracle2_parity = dj_oracle_circuit(2, func_type="balanced_parity")
dj2_parity = deutsch_jozsa(2, oracle2_parity)

# Custom 2-qubit oracle using the previously defined f_two_custom
oracle2_custom2 = dj_oracle_circuit(2, func_type="custom", custom_f=f_two_custom)
dj2_custom2 = deutsch_jozsa(2, oracle2_custom2)

display(Markdown("### Circuit (2-qubit DJ - Constant Zero)"))
display(dj2_const.draw(output='text'))

display(Markdown("### Circuit (2-qubit DJ - Balanced Parity)"))
display(dj2_parity.draw(output='text'))

display(Markdown("### Circuit (2-qubit DJ - Custom)"))
display(dj2_custom2.draw(output='text'))
```

```python
def deutsch_jozsa(n, oracle):
    """
    Build Deutsch-Jozsa circuit for n input qubits and a given oracle.
    """
    qc = QuantumCircuit(n + 1, n) # n input qubits, 1 output qubit, n classical bits
    qc.x(n) # Prepare ancilla qubit in |->
    qc.h(n)
    qc.h(range(n)) # Apply H gates to input qubits
    qc.barrier()
    qc.append(oracle.to_instruction(), range(n+1)) # Apply the oracle
    qc.barrier()
    qc.h(range(n)) # Apply H gates to input qubits
    qc.measure(range(n), range(n)) # Measure input qubits
    return qc
```

```python
display(Markdown("## Noise Impact on 2-qubit DJ (simple noise model)"))

noise_model = NoiseModel()
error_1 = depolarizing_error(0.01, 1)
error_2 = depolarizing_error(0.03, 2)
error_3 = depolarizing_error(0.03, 3) # Define a 3-qubit error
noise_model.add_all_qubit_quantum_error(error_1, ['h', 'x'])
noise_model.add_all_qubit_quantum_error(error_2, ['cx']) # Apply 2-qubit error to 'cx'
noise_model.add_all_qubit_quantum_error(error_3, ['ccx', 'mcx']) # Apply 3-qubit error to 'ccx', 'mcx'
read_err = ReadoutError([[0.98, 0.02], [0.03, 0.97]])
noise_model.add_all_qubit_readout_error(read_err)

sim = AerSimulator(noise_model=noise_model)
shots = 4000

for name, circ in [("constant_zero", dj2_const), ("balanced_parity", dj2_parity), ("custom_depends_on_bit0", dj2_custom2)]:
    t = transpile(circ, sim)
    res = sim.run(t, shots=shots).result()
    counts = res.get_counts()
    display(Markdown(f"### Noisy results for {name} (shots={shots})"))
    display(plot_histogram(counts))
    print(counts)
    p_all_zero = counts.get('00', 0)/shots
    display(Markdown(f"Probability of measuring all zeros: {p_all_zero:.3f}"))
```

```python
def classical_constant_zero(x, n):
    """Simulates a constant zero oracle classically."""
    return 0

def classical_constant_one(x, n):
    """Simulates a constant one oracle classically."""
    return 1

def classical_balanced_parity(x, n):
    """
    Simulates a balanced oracle based on parity classically.
    Calculates the parity of the bits in the integer x.
    """
    parity = 0
    for i in range(n):
        # Check the i-th bit of x
        if (x >> i) & 1:
            parity ^= 1 # XOR with 1 to flip parity if the bit is 1
    return parity
```

```python
def simulate_classical_dj(n, f, is_constant):
    """
    Simulate classical queries to distinguish between constant and balanced functions.

    Args:
        n: The number of input qubits (and classical bits).
        f: The classical function to evaluate (takes integer x and n as input).
        is_constant: True if the function is known to be constant, False if balanced.

    Returns:
        The number of queries needed to determine the function type.
    """
    queries = 0
    first_result = None

    # Iterate through possible inputs
    for x in range(2**n):
        queries += 1
        current_result = f(x, n)

        if first_result is None:
            first_result = current_result
        elif current_result != first_result:
            # Found a different result, function is balanced
            return queries

    # If the loop finishes without finding a different result
    if is_constant:
        return queries
    else:
        # This case should ideally not happen for truly balanced functions
        # that are evaluated over all possible inputs.
        print(f"Warning: Balanced function did not show difference after {queries} queries.")
        return queries

# Call the simulation function for the constant zero case
queries_const = simulate_classical_dj(2, classical_constant_zero, is_constant=True)

# Call the simulation function for the balanced parity case
queries_balanced = simulate_classical_dj(2, classical_balanced_parity, is_constant=False)

# Print the results
display(Markdown("## Classical Queries Needed"))
print(f"Queries needed for constant function: {queries_const}")
print(f"Queries needed for balanced function: {queries_balanced}")
```

```python
display(Markdown("## Comparing Classical and Quantum Queries"))

print(f"Classical queries needed to confirm a Constant function (worst case for n=2): {queries_const}")
print(f"Classical queries needed to detect a Balanced function (best/average case for n=2): {queries_balanced}")

display(Markdown(f"Quantum queries needed for any Deutsch-Jozsa function (constant or balanced, for any n): 1"))

display(Markdown(
"""
### Advantage of the Deutsch-Jozsa Algorithm

The Deutsch-Jozsa algorithm provides a dramatic speedup compared to classical computation for the specific task of determining whether a function is constant or balanced.

Classically, to definitively prove a function is constant, you must query it for every possible input (2^n queries in the worst case). To prove it's balanced, you might find two different outputs quickly (as in our simulation), but in the worst case for a balanced function that looks constant for many inputs, you might still need to query a significant number of inputs.

In contrast, the Deutsch-Jozsa quantum algorithm can determine whether the function is constant or balanced with just **one** query to the quantum oracle, regardless of the number of input qubits, *n*. This represents an exponential speedup for this particular problem.

While the ideal quantum algorithm is deterministic, real-world quantum computers are subject to noise. As seen in the noisy simulation results above, noise can introduce errors, leading to a non-zero probability of measuring an outcome that incorrectly suggests the function type. For instance, even for the 'constant zero' function, noise might cause some measurements to be non-zero, making the result less certain compared to an ideal classical computation. Mitigating these noise effects is an active area of research in quantum computing.
"""
))
```
